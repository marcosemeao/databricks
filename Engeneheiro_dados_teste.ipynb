```python
# Função para leitura do arquivo segmentando o dataset por chunks. Parâmetro chunksize
```

```python
def leitura_dado(dados, sep=';', chunk=1000000):
    return pd.read_csv(dados, sep=sep, chunksize=chunk)
```

```python
# Carregando os dados e inicializando o DataFrame intermediário que será usado como área temporária para acumular os dados capturados por chunks
```

```python
reader = leitura_dado(dados, sep, chunksize)
df = pd.DataFrame()
memory = pd.DataFrame()
```

```python
# Processando cada chunk de dados e realizando as operações de cálculo e agregação
```

```python
for chunk in reader:
    memory=pd.concat([memory,chunk.memory_usage()])
    temp = chunk.groupby(by=['Sales Channel','Item Type']).agg({'Units Sold':'sum'})
    df=pd.concat([df,temp])
```

```python
# Carregando os dados e inicializando o DataFrame intermediário que será usado como área temporária para acumular os dados capturados por chunks
```

```python
reader = leitura_dado(dados, sep=sep,chunk=chunksize)
df = pd.DataFrame()
memory = pd.DataFrame()
```

```python
# Processando cada chunk de dados e realizando as operações de cálculo e agregação
# Conversão do formato da data, somatoria e contagem dos valores das receitas para o cálculo da média, 
## os valores dos chunks são acumulados na camada intermediária
```

```python
for chunk in reader:
    memory=pd.concat([memory,chunk.memory_usage()])
    chunk['OrderDate'] = pd.to_datetime(chunk['Order Date'], format='%m/%d/%Y')
    temp=chunk.groupby(by=[chunk.OrderDate.dt.year,chunk.OrderDate.dt.month,'Item Type']).agg({'Total Revenue':['count','sum']})
    df=pd.concat([df, temp])
```

```python
# Medindo quantidade total de memória gasto no processamento em Bytes
```

```python
memory.sum(0)
```

```python
df
```

```python
# Renomeado o indice e as variáveis agregadas 
```

```python
df.index.names = ['OrderYear','OrderMonth','Item Type']
df.columns = ['RevenueCount','RevenueSum']
```

```python
df.memory_usage().sum()
```

```python
df
```

```python
# Consolidando os valores de receita e quantidade de ordens, agrupando por ano, mês e item do produto
```

```python
df1=df.groupby(by=['OrderYear','OrderMonth','Item Type']).agg({'RevenueCount':'sum','RevenueSum':'sum'})
```

```python
# Calculando a média do Volume de vendas por período (Ano,Mês)
```

```python
df1['RevenueAvg'] = df1['RevenueSum']/df1['RevenueCount']
```

```python
# Medindo quantidade total de memória gasto no processamento em Bytes
```

```python
df1.memory_usage().sum()
```

```python
# Ordenando os dados por Ano, Mês, Item do Produto e Valor médio da Receita em ordem ascendente, deixando os maiores valores no topo
```

```python
df2=df1.sort_values(by=['OrderYear','OrderMonth','RevenueAvg','Item Type'], ascending=False)
```
